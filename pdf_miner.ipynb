```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar

for page_layout in extract_pages("test.pdf"):
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            for text_line in element:
                font_sizes = []
                line_text = ""
                for character in text_line:
                    if isinstance(character, LTChar):
                        font_sizes.append(character.size)
                        line_text += character.get_text()
                if line_text.strip():
                    size_moyenne = sum(font_sizes)/len(font_sizes) if font_sizes else 0
                    print(f"[size={size_moyenne:.1f}] {line_text.strip()}")
```
